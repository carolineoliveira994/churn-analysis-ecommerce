```python
purchases = df[df['event_type'] == 'purchase']
```

```python
last_purchase = purchases.groupby('user_id')['time_id'].max()
```

```python
reference_time = df['time_id'].max()
```

```python
last_purchase = purchases.groupby('user_id')['time_id'].max()
reference_time = df['time_id'].max()
```

```python
churn_threshold = 30  # você pode ajustar

churn = (reference_time - last_purchase) > churn_threshold
```

```python
user_df = df.groupby('user_id').agg({
    'price': 'mean',
    'event_type': 'count'
}).rename(columns={
    'price': 'avg_price',
    'event_type': 'total_events'
})
```

```python
user_df['last_purchase'] = last_purchase
user_df['churn'] = churn
```

```python
user_df['churn'].mean()
```

```python
user_df.groupby('churn').mean()
```

```python
first_event = df.groupby('user_id')['time_id'].min()
lifetime = last_purchase - first_event

user_df['lifetime'] = lifetime
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = user_df[['avg_price', 'total_events', 'lifetime']].fillna(0)
y = user_df['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = RandomForestClassifier()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```
